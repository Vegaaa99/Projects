```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
C_1 = np.linspace(20,720,100) #Curva 1 a 720
C_2 = np.linspace(20,600,100) #Curva 2 a 600
T_1 = C_1/120  # Tiempo 1
T_2 = C_2/120  # Tiempo 2
T_1
```

```python
#Tiempo estable
E_1 = np.ones(100)*720
E_2 = np.ones(100)*600
TE_1 = np.linspace(T_1[-1],T_1[-1]+2,100)
TE_2 = np.linspace(T_2[-1],T_2[-1]+2.0,100)
TE_2
```

```python
plt.plot (T_1, C_1,color = 'red', label=r"Curva de $720^{\circ}C$")
plt.plot (T_2, C_2, color = 'orange', label=r"Curva de $600^{\circ}C$")
plt.plot (TE_1, E_1, color = 'red')
plt.plot (TE_2, E_2, color = 'orange' )
plt.xlabel(r'Tiempo [Horas]')
plt.ylabel(r'Temperatura $^{\circ}C$')
plt.title(r'Curvas de temperatura para $720^{\circ}C$ y $600^{\circ}C$')
plt.legend(loc='best')
plt.grid(True)
plt.savefig("Tarea1.pdf")
plt.savefig('curva de temperatura.png')
plt.show()
```
